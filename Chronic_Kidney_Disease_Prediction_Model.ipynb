## Chronic Kidney Disease Prediction Model Using Machine Learning

## Importing the Dataset 

```python
import pandas as pd
ckd_data = pd.read_csv('kidney_disease.csv')
ckd_data
```

```python
ckd_data.isnull().any()
```

```python
ckd_data.dropna(axis=0, inplace=True)
```

```python
#Converting the categorial labels into numerical
from sklearn.preprocessing import LabelEncoder
l_en = LabelEncoder()
```

```python
for i in [6,7,8,9,19,20,21,22,23,24]:
    ckd_data.iloc[:,i] = l_en.fit_transform(ckd_data.iloc[:,i])
```

```python
ckd_data
```

```python
ckd_data.describe()
```

```python
# assigning input, output variables
X = ckd_data.iloc[:,1:25].values
Y = ckd_data['classification'].values
X, Y
```

```python
# Splitting the dataset into train, test 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state=0)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
d_tree = DecisionTreeClassifier(criterion='gini')
```

```python
d_tree.fit(x_train, y_train)
```

```python
pred_d_tree = d_tree.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(pred_d_tree, y_test))
print(classification_report(pred_d_tree, y_test))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
print(confusion_matrix(lr_pred, y_test))
print(classification_report(lr_pred, y_test))
```

```python
accuracy_score(lr_pred, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
```

```python
gb.fit(x_train, y_train)
```

```python
gb_pred = gb.predict(x_test)
```

```python
print(confusion_matrix(gb_pred, y_test))
print(classification_report(gb_pred, y_test))
```

```python
accuracy_score(gb_pred, y_test)
```

```python
#K-nn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)
```

```python
print(confusion_matrix(knn_pred, y_test))
print(classification_report(knn_pred, y_test))
```

```python
#AdaBoost
from sklearn.ensemble import AdaBoostClassifier
adb = AdaBoostClassifier()
```

```python
adb.fit(x_train, y_train)
```

```python
adb_pred = adb.predict(x_test)
```

```python
print(confusion_matrix(adb_pred, y_test))
print(classification_report(adb_pred, y_test))
```

```python
accuracy_score(adb_pred, y_test)
```
